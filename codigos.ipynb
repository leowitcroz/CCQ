```python
from qiskit import QuantumCircuit, Aer, transpile
from qiskit.visualization import plot_histogram
from math import sqrt
import matplotlib.pyplot as plt
def teleportar_estado(alpha, beta):
    qc = QuantumCircuit(3, 2)
    # Qubit 0: A1- estado a ser teleportado
    # Qubit 1: A2- qubit de Alice, entrelacado com Bob
    # Qubit 2: B- qubit de Bob
    # 1. Inicializa o estado a ser teleportado em A1
    qc.initialize([alpha, beta], 0)
    # 2. Criar par emaranhado entre A2 e B
    qc.h(1)
    qc.cx(1, 2)
    # 3. Operacoes de Alice (A1 e A2)
    qc.cx(0, 1)
    qc.h(0)
    # 4. Medicoes de Alice
    qc.measure(0, 0)
    qc.measure(1, 1)
    # 5. Simulacao
    sim = Aer.get_backend('aer_simulator')
    qc = transpile(qc, sim)
    result = sim.run(qc, shots=1024).result()
    counts = result.get_counts()
    print("Resultado das medicoes de Alice:", counts)
    plot_histogram(counts)
    plt.show()
# Exemplo: teleportar (1/ 2 )|0> + (1/ 2 )|1>
teleportar_estado(1/sqrt(2), 1/sqrt(2))
```

```python
# Importando as bibliotecas necessarias
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt

# Funcao para Superdense Coding
def superdense_coding(bits: str):
    if bits not in ["00", "01", "10", "11"]:
        raise ValueError("Bits must be one of: '00', '01', '10', '11'")
    
    qc = QuantumCircuit(2, 2)
    
    # 1. Criar estado emaranhado (Bell state) entre Alice (q0) e Bob (q1)
    qc.h(0)
    qc.cx(0, 1)
    qc.barrier() # Barreira para visualização
    
    # 2. Codificacao dos bits por Alice no seu qubit (q0)
    # Para "00", nenhuma porta é aplicada.
    if bits == "01":
        qc.x(0)  # Aplica porta X para "01"
    elif bits == "10":
        qc.z(0)  # Aplica porta Z para "10"
    elif bits == "11":
        qc.z(0)  # Aplica porta Z
        qc.x(0)  # e depois porta X para "11"
        
    qc.barrier() # Barreira para visualização
        
    # 3. Decodificacao de Bob
    # Bob aplica as portas para reverter o emaranhamento
    qc.cx(0, 1)
    qc.h(0)
    
    # 4. Medidas
    # Bob mede os dois qubits para obter os bits clássicos
    qc.measure(0, 1) # Mede o qubit 0 no bit clássico 1
    qc.measure(1, 0) # Mede o qubit 1 no bit clássico 0
    
    # Simulador Aer
    simulator = AerSimulator()
    result = simulator.run(qc, shots=1024).result()
    counts = result.get_counts()
    
    print(f"Bits enviados por Alice: {bits}")
    print(f"Resultado da medicao de Bob: {counts}")
    
    # Exibe o circuito
    display(qc.draw('mpl'))
    # Exibe o histograma
    plot_histogram(counts)
    plt.show()

# Exemplo: enviar os bits "10"
superdense_coding("10")
```

```python
# 1. Importando as bibliotecas necessárias
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator # Importa o simulador do pacote Qiskit Aer
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# Função principal do Algoritmo de Deutsch
def deutsch_algorithm(U_f):
    # Inicializa um circuito com 2 qubits e 1 bit clássico para leitura
    qc = QuantumCircuit(2, 1)
    
    # Prepara o estado inicial |0>|1>
    qc.x(1) # Aplica X no segundo qubit para preparar |1>
    qc.barrier()
    
    # Aplica Hadamard em ambos os qubits para criar superposição
    qc.h(0)
    qc.h(1)
    qc.barrier()
    
    # Aplica o oráculo (caixa preta) Uf
    qc.append(U_f, [0,1])
    qc.barrier()
    
    # Aplica Hadamard no primeiro qubit para a interferência
    qc.h(0)
    
    # Mede o primeiro qubit
    qc.measure(0, 0)
    
    # Executa o circuito no simulador
    simulator = AerSimulator() # Cria uma instância do simulador
    result = simulator.run(qc, shots=1024).result() # Executa e obtém o resultado
    counts = result.get_counts()
    
    # Mostra o circuito e o resultado
    print("Circuito Quântico:")
    display(qc.draw('mpl')) # 'mpl' para uma visualização gráfica melhor
    print("\nResultados da Medição:", counts)
    plot_histogram(counts)
    plt.show()

# Define o oráculo Uf para a função constante f(x)=0
def constant_Uf():
    qc = QuantumCircuit(2)
    # Para f(x)=0, Uf é a identidade, não altera os qubits
    return qc.to_gate(label='Uf_constante')

# Define o oráculo Uf para a função balanceada f(x)=x
def balanced_Uf():
    qc = QuantumCircuit(2)
    # Para f(x)=x, Uf implementa CNOT: |x, y> -> |x, y XOR f(x)>
    qc.cx(0,1)
    return qc.to_gate(label='Uf_balanceada')

# --- Execução do Algoritmo ---

# 1. Testar com a função constante
print("="*40)
print("Executando para a função constante f(x)=0")
print("Resultado esperado: Medição sempre será '0'")
print("="*40)
deutsch_algorithm(constant_Uf())

# 2. Testar com a função balanceada
print("\n" + "="*40)
print("Executando para a função balanceada f(x)=x")
print("Resultado esperado: Medição sempre será '1'")
print("="*40)
deutsch_algorithm(balanced_Uf())
```

```python
# Importando as bibliotecas necessárias
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator # Importa o simulador do pacote Qiskit Aer
from qiskit.visualization import plot_histogram
import random
import numpy as np

def deutsch_jozsa_oracle(n, tipo='balanceado'):
    """
    Cria um oráculo para o algoritmo de Deutsch-Jozsa.
    - Constante: f(x) = 0 ou f(x) = 1 para todo x.
    - Balanceado: f(x) = s · x (mod 2), onde s é uma bitstring secreta.
    """
    oracle = QuantumCircuit(n + 1)
    
    if tipo == 'constante':
        # Decide aleatoriamente se f(x)=0 ou f(x)=1
        if random.choice([True, False]):
            # f(x) = 1. Aplica X no qubit de saída.
            oracle.x(n)
        # Se for f(x)=0, o oráculo é a identidade (não faz nada).
            
    elif tipo == 'balanceado':
        # Uma forma garantida de criar uma função balanceada é escolher uma
        # bitstring secreta 's' (não nula) e implementar f(x) = s · x (mod 2).
        # Aqui, vamos escolher uma 's' aleatória.
        s = np.random.randint(1, 2**n) # Gera um inteiro aleatório entre 1 e 2^n - 1
        s_bin = format(s, f'0{n}b') # Converte para binário com n bits
        
        # Aplica CNOT para cada '1' na bitstring secreta 's'
        for i, bit in enumerate(s_bin):
            if bit == '1':
                oracle.cx(i, n)
    else:
        raise ValueError("O tipo do oráculo deve ser 'constante' ou 'balanceado'")
        
    return oracle

def deutsch_jozsa_algorithm(n, tipo='balanceado'):
    """Executa o algoritmo Deutsch-Jozsa para n bits."""
    
    # Cria o circuito com n+1 qubits e n bits clássicos
    circuit = QuantumCircuit(n + 1, n)

    # 1. Prepara o estado inicial
    # Aplica Hadamard nos n primeiros qubits
    circuit.h(range(n))
    # Coloca o último qubit no estado |->
    circuit.x(n)
    circuit.h(n)
    circuit.barrier()

    # 2. Adiciona o oráculo
    oracle = deutsch_jozsa_oracle(n, tipo)
    oracle.name = f"Oráculo ({tipo})"
    circuit.append(oracle.to_gate(), range(n + 1))
    circuit.barrier()

    # 3. Aplica Hadamard nos primeiros n qubits novamente
    circuit.h(range(n))

    # 4. Mede os primeiros n qubits
    circuit.measure(range(n), range(n))

    # Executa o circuito usando o AerSimulator
    simulator = AerSimulator()
    # O algoritmo é determinístico, então 1 shot é suficiente.
    result = simulator.run(circuit, shots=1).result()
    counts = result.get_counts()
    
    # Retorna o primeiro (e único) resultado medido
    resultado_medido = list(counts.keys())[0]
    
    return circuit, resultado_medido

# --- Exemplo de uso ---
n = 4  # Número de qubits de entrada

# Teste com oráculo CONSTANTE
print("="*40)
print(f"Testando com oráculo CONSTANTE para {n} bits.")
print("Resultado esperado: '0000'")
circuito_const, resultado_const = deutsch_jozsa_algorithm(n, tipo='constante')
print(f"Resultado da medição: '{resultado_const}'")
print("="*40, "\n")
# display(circuito_const.draw('mpl')) # Descomente para ver o circuito

# Teste com oráculo BALANCEADO
print("="*40)
print(f"Testando com oráculo BALANCEADO para {n} bits.")
print("Resultado esperado: Qualquer bitstring diferente de '0000'")
circuito_bal, resultado_bal = deutsch_jozsa_algorithm(n, tipo='balanceado')
print(f"Resultado da medição: '{resultado_bal}'")
print("="*40)
# display(circuito_bal.draw('mpl')) # Descomente para ver o circuito
```

```python
import math
import random
# --- Parte Quântica (Usando API Antiga do Qiskit) ---
# ATENÇÃO: A classe 'Shor' foi removida do Qiskit 1.0 e versões posteriores.
# Este código só funcionará com uma versão mais antiga do Qiskit instalada.
# Ex: pip install qiskit==0.45 qiskit-aer==0.12
from qiskit_aer import Aer
from qiskit.utils import QuantumInstance
from qiskit.algorithms import Shor

def shor_algorithm_qiskit(N, a=None):
    """
    Executa o algoritmo de Shor usando a classe de alto nível (obsoleta) do Qiskit.
    Requer uma versão antiga do Qiskit para funcionar.
    """
    print("Executando a fatoração com a classe Shor (obsoleta) do Qiskit...")
    # Configura o backend quântico (simulador)
    # QuantumInstance também é uma classe obsoleta
    backend = Aer.get_backend('qasm_simulator')
    quantum_instance = QuantumInstance(backend, shots=1024)
    
    # Execução do algoritmo de Shor
    shor = Shor(quantum_instance=quantum_instance)
    result = shor.factor(N, a=a)
    
    return result

# --- Parte Clássica (Simulação da Lógica Matemática) ---
def shor_classical_simulation(N):
    """
    Simula os passos matemáticos clássicos do algoritmo de Shor
    para encontrar os fatores de N.
    """
    print("\nExecutando a simulação clássica da lógica de Shor...")
    # Passo 1: Lidar com casos triviais
    if N % 2 == 0:
        return (2, N // 2)
    if math.isqrt(N)**2 == N:
        return (math.isqrt(N), math.isqrt(N))

    # Passo 2: Loop principal para encontrar os fatores
    while True:
        # a) Escolher um número aleatório 'a'
        a = random.randint(2, N - 1)
        
        # b) Verificar se 'a' já é um fator
        k = math.gcd(a, N)
        if k != 1:
            return (k, N // k)

        # c) Encontrar o período 'r' (esta é a parte difícil que um computador quântico acelera)
        # Classicamente, fazemos isso por força bruta
        for r in range(1, N):
            if pow(a, r, N) == 1:
                break # Período 'r' encontrado
        
        # d) Se o período 'r' for ímpar ou levar a um resultado trivial, tente outro 'a'
        if r % 2 != 0:
            continue
        
        y = pow(a, r // 2, N)
        if y == N - 1:
            continue
            
        # e) Encontrar os fatores usando o máximo divisor comum (MDC)
        p = math.gcd(y - 1, N)
        q = math.gcd(y + 1, N)

        # Retorna os fatores não triviais encontrados
        return (p, q)

# --- Exemplo de Uso ---
N = 15

# 1. Execução com a simulação clássica (funciona em qualquer versão do Qiskit)
fatores_classicos = shor_classical_simulation(N)
print(f"Fatores de {N} (simulação clássica): {fatores_classicos}")

# 2. Execução com a versão do Qiskit (requer versão antiga)
try:
    result = shor_algorithm_qiskit(N)
    print(f"Fatores de {N} (Qiskit obsoleto): {result.factors[0]}")
except (ImportError, AttributeError):
    print("\nNão foi possível executar a versão quântica.")
    print("Motivo: A classe 'qiskit.algorithms.Shor' foi removida do Qiskit 1.0+.")
```
